```python
from pathlib import Path
import numpy as np
from scipy.io import wavfile
from tqdm import tqdm
```

```python
versions = [1, 5, 8, 9, 13, 14, 16, 18]
epochs = ["014", "029", "044"]
```

```python
data_dir = Path("/run/media/nathan/Nathan4T/Nathan-canarygan/generation")
```

```python
save_dirs = {
    "014": Path("/run/media/nathan/Nathan4T/Nathan-canarygan/data/decoder/EARLY_15"),
    "029": Path("/run/media/nathan/Nathan4T/Nathan-canarygan/data/decoder/EARLY_30"),
    "044": Path("/run/media/nathan/Nathan4T/Nathan-canarygan/data/decoder/EARLY_45"),
}

for d in save_dirs.values():
    d.mkdir(parents=True, exist_ok=True)
```

```python
files = []
for e in epochs:
    i = 0
    for v in versions:
        f = list((data_dir / f"version_{v}").glob(f"*epoch_{e}*"))[0]
        d = np.load(f)
        x = d["x"]
        idxs = np.random.choice(np.arange(len(x)), size=1000//len(versions), replace=False)
        sel = x[idxs]
        for xx in tqdm(sel):
            wavfile.write(save_dirs[e] / f"early_{int(e) + 1}-{i}.wav", rate=16000, data=xx)
            i += 1
```
